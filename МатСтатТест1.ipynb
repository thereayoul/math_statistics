```python
n = 29
score = [90, 79, 53, 62, 66, 68, 75, 0, 82, 29, 0, 29, 68, 90, 0, 60, 44, 44, 70, 68, 70, 89, 0, 68, 0, 66, 0, 59, 70]

import numpy as np
import statistics

A = statistics.mean([i for i in score if i>0])
print("A:", str(round(A,3)).replace('.', ','))

M = statistics.median([i for i in score if i>0])
print("M:",str(round(M,2)).replace('.', ','))

H = statistics.harmonic_mean([i for i in score if i>=M])
print("H:",str(round(H,3)).replace('.', ','))

#      geometric_mean...
#G = statistics.geometric_mean([i for i in score if i>=M])

#  
import numpy as np
def geo_mean(iterable):
    a = np.array(iterable)
    return a.prod()**(1.0/len(a))

G = geo_mean([i for i in score if i>=M])
print(str(round(G,3)).replace('.', ','))


Q = statistics.median([i for i in score if i>=M])
print("Q:",str(round(Q,2)).replace('.', ','))

# N
```

```python
from statistics import mean
from numpy import std
from statistics import NormalDist
import scipy.stats as sps
#   3  
pd = [9, 58, 310, -686, 376, -71, 117, -349, 480, -151, 269, 6, -248, -104, -220, 233, 37, -324, -106, 230]
one = mean(pd)
two = std(pd)
norm = NormalDist(one, two)
L = norm.quantiles(n=4)[0]
H = norm.quantiles(n=4)[2]
four = len([x for x in pd if L <= x <= H])
print(one)
print(two)
print(L)
print(H)
print(four)
print(pd.cdf(2))
```

```python
k = 3
n = [24, 26, 30]
xi = [70, 76, 77]
sigma = [4, 6, 8]


N = sum(n)

a = []
for i in range(len(n)):
  a.append(n[i]*xi[i])
x = 1/N *sum(a)
print(" :",str(round(x,3)).replace('.', ','))

b = []
for i in range(k):
  b.append(n[i]*(xi[i]-x)**2)

c = []
for i in range(k):
  c.append(sigma[i]**2*n[i])

sd = (1/N * (sum(b) + sum(c)))**0.5
print(" :",str(round(sd,3)).replace('.', ','))
```
